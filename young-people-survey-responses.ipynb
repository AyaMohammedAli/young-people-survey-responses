```python
import seaborn as sns 
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x = 'total_bill', y ='tip', data = tips,hue ='smoker')
plt.show()
```

```python
sns.countplot(x = 'smoker', data = tips)
plt.show()
```

```python
df2 = pd.read_csv('student-alcohol-consumption.csv')
df2.head()
```

```python
sns.scatterplot(x ='failures',y = 'absences',data = df2 )
plt.show()
```

```python
sns.relplot(x="absences", y="G3", 
            data=df2,
            kind="scatter",row ='study_time'
            )

# Show plot
plt.show()
```

```python
sns.relplot(x="absences", y="G3", 
            data=df2,
            kind="scatter",col ='study_time'
            )

# Show plot
plt.show()
```

```python
sns.relplot(x="G1", y="G3", 
            data=df2,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row ='famsup',
            row_order=["yes", "no"])

# Show plot
plt.show()
```

```python
df3 = pd.read_csv('mpg.csv')
df3
                  
```

```python
sns.relplot(x ="horsepower",y = "mpg",kind ='scatter',size =  'cylinders',data = df3,hue = 'cylinders')


# Show plot
plt.show()
```

```python
sns.relplot(kind ='scatter',data = df3,x = "acceleration" ,y = "mpg",style = "origin",hue ='origin' )
plt.show()
```

*cars from usa use mpg less and more acceleration rather than cars in japan and europe*

```python
sns.relplot(x="model_year", y="horsepower", 
            data=df3, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers=True,
            dashes=False)

# Show plot
plt.show()
```
